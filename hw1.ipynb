### **Part 1: Data Cleaning and Exploration**

##### 1. Load the dataset. Check the first 5 rows.

```python
import pandas as pd

# Load the dataset
crime_data = pd.read_csv("crime_data.csv")

# Show the first 5 rows
crime_data.head()
```

##### 2. Identify columns with missing values and their respective counts. Drop columns where more than 50% of the data is missing (store this version as a new dataset).

```python
# Check missing values per column
missing_data = crime_data.isnull().sum()
totalRows = len(crime_data)

# Drop columns with more than 50% missing values
threshold = totalRows * 0.5

columns_to_drop = missing_data[missing_data > threshold]
# Print the dropped columns and their missing data count
print("Dropped columns and their missing data count:")
print(columns_to_drop)

cleaned_data = crime_data.dropna(axis=1, thresh=threshold)
cleaned_data.head()
```

##### 3. Convert the DATE OCC column to a datetime format. Extract the year, month, and day into separate columns. Create a new column for the hour using the TIME OCC column.

```python
# Ensure that cleaned_data is a copy to avoid a SettingWithCopyWarning
cleaned_data = cleaned_data.copy()

# Convert DATE OCC to datetime format
cleaned_data['DATE OCC'] = pd.to_datetime(cleaned_data['DATE OCC'], errors='coerce')

# Extract year, month, day
cleaned_data['Year'] = cleaned_data['DATE OCC'].dt.year
cleaned_data['Month'] = cleaned_data['DATE OCC'].dt.month
cleaned_data['Day'] = cleaned_data['DATE OCC'].dt.day

# Extract hour from TIME OCC (assuming TIME OCC is in HHMM format)
cleaned_data['Hour'] = cleaned_data['TIME OCC'].astype(str).str.zfill(4).str[:2].astype(int)

# Display the first few rows to verify
cleaned_data.head()
```

##### 4. Filter the dataset for crimes that occurred in 2023. Further filter crimes with the description BURGLARY in the Crm Cd Desc column.

```python
#Filtering by year
crime_data_2023 = cleaned_data[(cleaned_data['Year'] == 2023)]
#Filtering further by Crm Cd Desc
crime_data_2023_burglary = crime_data_2023[(crime_data_2023['Crm Cd Desc'] == 'BURGLARY')]
print('Clean Data:', len(cleaned_data))
print('Only 2023 crimes:', len(crime_data_2023))
print('Only 2023 crimes that are burglary:', len(crime_data_2023_burglary))
crime_data_2023_burglary.head()
```

##### 5. Group the data by AREA NAME and calculate the total number of crimes and the average victim age. Sort the results by total crimes in descending order.

```python
# Group by AREA NAME, calculate total crimes and average victim age
grouped_data = cleaned_data.groupby('AREA NAME').agg(
    total_crimes=('Crm Cd Desc', 'count'),
    avg_victim_age=('Vict Age', 'mean')
).sort_values(by='total_crimes', ascending=False)

grouped_data.head(len(grouped_data))
```

### **Part 3: Advanced Analysis**

##### Create a new column, Severity Score, based on the following rules:
- Assign 5 points if a weapon was used (Weapon Used Cd is not null).
- Assign 3 points for crimes under BURGLARY.
- Assign 1 point for all other crimes.
- Group by AREA NAME and find the total severity score for each area.

```python
# Create Severity Score based on rules
def severity_score(row):
    if pd.notnull(row['Weapon Used Cd']):
        return 5
    elif row['Crm Cd Desc'] == 'BURGLARY':
        return 3
    else:
        return 1

crime_data['Severity Score'] = crime_data.apply(severity_score, axis=1)

# Group by AREA NAME and find total severity score
severity_by_area = crime_data.groupby('AREA NAME')['Severity Score'].sum()

severity_by_area.head(len(severity_by_area))
```

##### Use the LAT and LON columns to identify crimes that occurred within a specific latitude-longitude bounding box (e.g., downtown area).

- North Boundary: North of 3rd Street
- South Boundary: South of 7th Street
- East Boundary: East of Alameda Street
- West Boundary: West of the 110 Freeway

```python
# Define bounding box (latitude and longitude)
lat_min, lat_max = 34.04, 34.05
lon_min, lon_max = -118.26, -118.25

# Filter crimes within bounding box
filtered_by_location = cleaned_data[(crime_data['LAT'] >= lat_min) & 
                                  (crime_data['LAT'] <= lat_max) &
                                  (crime_data['LON'] >= lon_min) &
                                  (crime_data['LON'] <= lon_max)]
print('Total crimes in downtown area:', len(filtered_by_location))
filtered_by_location.head()
```
